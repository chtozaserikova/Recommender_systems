```python
import pandas as pd
metadata = pd.read_csv('dataset/movies.csv', low_memory=False)
metadata.head(3)
```

```python
C = metadata['score'].mean()
print(C)
m = metadata['votes'].quantile(0.90)
print(m)
```

```python
q_movies = metadata.copy().loc[metadata['votes'] >= m]
print(q_movies.shape)
print(metadata.shape)
```

Из приведенного выше вывода становится ясно, что около 10% фильмов имеют более 226000 голосов и могут быть включены в этот список.

```python
def weighted_rating(x, m=m, C=C):
    v = x['votes']
    R = x['score']
    # Расчет по формуле IMDB
    return (v/(v+m) * R) + (m/(m+v) * C)

q_movies['new_score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)

#топ 10 фильмов
q_movies[['name', 'votes', 'score', 'new_score']].head(10)
```
